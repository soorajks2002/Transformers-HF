```python
from transformers import pipeline
```

```python
ner_model = pipeline("token-classification", model="dslim/bert-base-NER")
```

```python
text = "Srinivasa Ramanujan was an Indian mathematician who lived from 1887 to 1920. He had almost no formal training in pure mathematics, but he made substantial contributions to mathematical analysis, number theory, infinite series, and continued fractions, including solutions to mathematical problems then considered unsolvable."
```

```python
output = ner_model(text)
output
```

```python
for entity in output :
  print("Entity : ", entity['entity'])
  print("Word : ", entity['word'])
  print("\n")
```
