# Using HuggingFace Pipeline

```python
from transformers import pipeline,set_seed
```

```python
pipe = pipeline("text-generation", model="gpt2")
```

```python
set_seed(24)
```

```python
pipe("this is an AI named sevak", max_length=50, num_return_sequences=2)
```

# Using HuggingFace Tokenizer & Model

```python
import tensorflow as tf
from transformers import GPT2Tokenizer, TFGPT2LMHeadModel
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium')
model = TFGPT2LMHeadModel.from_pretrained('gpt2-medium')
```

```python
encoded_input = tokenizer.encode("this is an AI named sevak", add_special_tokens=True)
input_ids = tf.constant(encoded_input)[None, :]
output = model.generate(input_ids=input_ids)
```
