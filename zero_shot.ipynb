```python
from transformers import pipeline
```

```python
pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "Srinivasa Ramanujan was an Indian mathematician who lived from 1887 to 1920. He had almost no formal training in pure mathematics, but he made substantial contributions to mathematical analysis, number theory, infinite series, and continued fractions, including solutions to mathematical problems then considered unsolvable."
candidate_labels = ['India', 'Maths', 'Cooking', 'Electronics']
```

```python
pipe(sequence_to_classify, candidate_labels)
```

```python
pipe(sequences = sequence_to_classify, candidate_labels = candidate_labels)
```
